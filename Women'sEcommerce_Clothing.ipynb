```python
tokens_final=[]
for i in tokens:
    new=[]
    for j in i:
        if not(j in final_stop):
            new.append(j)
    tokens_final.append(new)
```

```python
from nltk.stem import PorterStemmer

#from nltk.tokenize import sent_tokenize,word_tokenize

ps=PorterStemmer()

stemmed_words=[]

for w in tokens_final:
    new=[]
    for j in w:
        new.append(ps.stem(j))
    #print(new)
    stemmed_words.append(new)
```

```python
stemmed_words[92]
```

```python
final_text=[]
for i in stemmed_words:
    new=[]
    for j in i:
        new.append(j)
    final_text.append(' '.join(new))
```

```python
final_text[0]
```

```python
type(final_text)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( final_text, data['Recommended IND'], 
                                                    test_size = 0.2, random_state = 42)
```

```python
X_train[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect=TfidfVectorizer(min_df=5).fit(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn import metrics
```

```python
X_train_vectorized=vect.transform(X_train)

model=LogisticRegression()
model.fit(X_train_vectorized,y_train)

predictions=model.predict(vect.transform(X_test))

print("AUC for Tfidf",metrics.roc_auc_score(y_test,predictions))
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB().fit(X_train_vectorized,y_train)
predicted= clf.predict(vect.transform(X_test))
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Fit the CountVectorizer to the training data specifiying a minimum 
# document frequency of 5 and extracting 1-grams and 2-grams
vect = CountVectorizer(min_df=5, ngram_range=(1,2)).fit(X_train)

X_train_vectorized = vect.transform(X_train)

#len(vect.get_feature_names())
```

```python
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)

predictions = model.predict(vect.transform(X_test))

print('AUC for n-grams: ', metrics.roc_auc_score(y_test, predictions))
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr,tpr,thr=roc_curve(y_test,predictions)
```

```python
y_prob = model.predict_proba(vect.transform(X_test))
```

```python
y_prob
```

```python
prob=[]
for i in y_prob:
    prob.append(i[1])
```

```python
fpr,tpr,thr=roc_curve(y_test,prob,pos_label=1)
```

```python
Pref=pd.DataFrame({'FPR':fpr,'TPR':tpr,'Threshold':thr})
```

```python
Pref['F1']=2*(1-Pref['FPR'])*(Pref['TPR'])/(Pref['TPR'] + 1-Pref['FPR'])
```

```python
Pref.head()
```

```python
Pref[Pref['F1'] == np.max(Pref['F1'].values)]
```
